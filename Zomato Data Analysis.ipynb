## Zomato data analysis project

Pandas is used for data manipulation and analysis.
numpy is used for numerical operations.
matplotlib.pyplot as seaborn used for data visualiztion.

Step 1- importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib. pyplot as plt
import seaborn as sns
```

Step  - Create a dataframe

```python
dataframe = pd.read_csv('/content/Zomato data .csv')
print(dataframe)
```

convert the data type of column - rate

```python
def handleRate(value):
    value = str(value).split('/')
    value = value[0]
    return float(value)

dataframe['rate'] = dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe.info()
```

## Type of resturant

```python
dataframe.head()
```

```python
sns.countplot(x = dataframe['listed_in(type)'])
plt.xlabel("type of resturannt") # Changed 'xlable' to 'xlabel'
```

## conclusion - majority of the resturant falls in dining category

```python
dataframe.head()
```

```python
Group_data = dataframe.groupby('listed_in(type)')['votes'].sum() # This line calculates the sum of 'votes' for each group.
result = pd.DataFrame({'votes':Group_data})
plt.plot(result,c="green") #removed extra color and invalid size
plt.xlabel("type of resturant", c="red", size =20) # Changed 'xlable' to 'xlabel'
plt.ylabel("votes",c="red",size=20)
```

## conclusion - dinning resturants has recieved maximum votes

```python
dataframe.head()
```

```python
plt.hist(dataframe['rate'],bins = 10)
plt.title("ratings distribution")
plt.show()
```

## conclusion - the majority resturants recieved rating from 3.5 to 4

## average order spending by couples

```python
dataframe.head()
```

```python
couple_data = dataframe['approx_cost(for two people)']
sns.countplot(x = couple_data)
```

## conclusion - the majority of couples preffer resutarnts with an approximate cost of 300 rupees

## which mode recieved maxiimum rating

```python
plt.figure(figsize = (6,6))
sns.boxplot(x = 'online_order', y = 'rate', data = dataframe)
```

##conclusion - offline order recieved lower rating in comparision to online order

```python
dataframe.head()
```

```python
pivot_table = dataframe.pivot_table(index='listed_in(type)',columns = 'online_order', aggfunc= 'size',fill_value=0) # Assign the result to pivot_table
sns.heatmap(pivot_table,annot=True,cmap= "YlGnBu",fmt='d') # Correct annot and cmap parameters
plt.title("heatmap")
plt.xlabel("online order") # Correct xlabel
plt.ylabel("listed in (type)")
plt.show()
```

#conclusion - Dining restaurants primarily accept offline orders, whereas cafes primarily receive online orders. This suggests that clients prefer to place orders in person at restaurants but prefer online ordering at cafes
